```python
import numpy as np
import matplotlib.pyplot as plt
import math

h=6.62*10**-34 #SI unit
c=3*10**8      #speed of light
k=1.38*10**-23 #Boltzmann Constant JK^-1

v=np.linspace(10**0, 10**15.5, 100, endpoint=True)
T=5000
#T=float(input("Input Temperature"))
B=2*h*v**3/c**2/(math.e**(h*v/k/T)-1)
num=100

#plt.errorbar(v, B, yerr=0.5*B, xerr=0.1*v, ecolor="green")

plt.plot(v, B, "r--")
#plt.xlim(10**14, 10**15.2)
#plt.ylim(10**-8, 10**-9)
#plt.loglog(v, B)  #log
plt.xlabel("Frequency (Hz)")
plt.ylabel("Intensity Watts/Hz/m^2")

dB=np.random.normal(0., 10**-8.5,num)
B += dB

plt.plot(v,B)
#plt.loglog(v,B)

plt.show()
```
